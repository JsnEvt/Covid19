```python
#a coluna UF na tabela abaixo nao permite ser usada para montar o grafico ja que trata-se do indice.
maximo
```

```python
maximo.index.name = 'UF'
```

```python
maximo
```

```python
#criando uma coluna UF para servir como referencia para montagem do grafico, passando como parametro, o proprio indice.
maximo_uf = maximo.assign(UF=maximo.index)
```

```python
maximo_uf
```

```python
maximo_uf.info()
```

```python
#usando a sns.barplot para montagem do grafico ja que nao consegui com o countplot. diferenca = usamos x e y como parametros.
plt.figure(figsize=(12,6))
#sns.countplot(x='UF', data=maximo_uf)
sns.barplot(x='UF', y='deaths', data=maximo_uf)
```

```python
#EXTRAIR AS COLUNAS DE STATE, CASES, DEATHS E AGRUPA-LAS PARA ANALISE
```

```python
#agrupando dia 23/05/2021
df_agrupado_data = df[df['date']=='2021-05-23']
df_agrupado_data
```

```python
df_agrupado_data = df_agrupado_data.drop('code', axis=1)
```

```python
df_agrupado_data
```

```python
df_agrupado_data = df_agrupado_data.drop('name', axis=1)
```

```python
df_agrupado_data
```

```python
soma_df_agrupado_data = df_agrupado_data.groupby('state').sum()
```

```python
soma_df_agrupado_data
```

```python
#COMO CRIAR UMA COLUNA QUE FACA A CORRELACAO EM PERCENTUAL ENTRE CASOS E MORTES POR ESTADO
soma_df_agrupado_data = soma_df_agrupado_data.assign(UF=soma_df_agrupado_data.index)
```

```python
soma_df_agrupado_data
```

```python
#adicionando uma coluna para colocar o percentual
coluna_perc = soma_df_agrupado_data.assign(Percentual='')
```

```python
coluna_perc.head()
```

```python
casos = coluna_perc['cases']
```

```python
casos.dtype
```

```python
lista_casos = np.array(casos)
```

```python
lista_casos
```

```python
lista_casos.dtype
```

```python
mortes = coluna_perc['deaths']
```

```python
lista_mortes = np.array(mortes)
```

```python
lista_mortes
```

```python
#criando a funcao que fara o calculo do percentual
def perc(x):
    for i in lista_casos:
        x = (lista_mortes / lista_casos)*100
        return x
```

```python
coluna_percentual = coluna_perc['Percentual'].apply(lambda x: perc(x))
```

```python
#para cada linha ele calculou todos os estados ....
coluna_percentual.head()
```
